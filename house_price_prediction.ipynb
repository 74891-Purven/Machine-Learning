# Preprocessing

- when features are highly skewed  then we convert them to normal distribution using power transform

power transform types

- log transform
- box-cox transform
- Yeo-Johnson Transform


here we are going to transform target column (price) , sqft_living, sqft_lot, sqft_living15,sqft_lot15

```python
df.describe()
```

```python
df.skew()
```

```python
import numpy as np
```

```python
# based on skewness on numeric columns we decide to transform following columns
df['price'] = np.log(df['price'])
df['sqft_lot'] = np.log(df['sqft_lot'])
df['sqft_lot15'] = np.log(df['sqft_lot15'])
```

```python
df.describe()
```

```python
df.skew()
```

# Seperate X & Y

Y --> price

X --> everything else

```python
X = df.drop('price',axis=1)
Y = df['price']
```

```python
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=7)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
ls = Lasso(alpha=1.0,random_state=7)
```

```python
ls.fit(X_train,Y_train)
```

```python
for col, coef in zip(X_train.columns,ls.coef_):
    print(col,coef)
```

```python
ls = Lasso(alpha=0.01,random_state=7,max_iter=5000 )
ls.fit(X_train,Y_train)
for col, coef in zip(X_train.columns,ls.coef_):
    print(col,coef)
```

```python
# Higher value of alpha ---- more features will have 0 coefficients so more features are dropped!!
ls = Lasso(alpha=0.5,random_state=7,max_iter=5000 )
ls.fit(X_train,Y_train)
for col, coef in zip(X_train.columns,ls.coef_):
    print(col,coef)
```

# How to fine tune alpha

-- Use for loop

-- Use grid search

```python
from sklearn.metrics import r2_score
```

```python
alpha_list = [0.01,0.05,0.1,0.5,1]
r2_score_list=[]
for alpha_val in alpha_list:
    ls = Lasso(alpha=alpha_val,random_state=7,max_iter=5000)
    ls.fit(X_train,Y_train)
    Y_pred = ls.predict(X_test)
    r2_score_list.append(r2_score(Y_test,Y_pred))
```

```python
r2_score_list
```

# Ridge Model

```python
from sklearn.linear_model import Ridge
```

```python
rg = Ridge(alpha = 1, max_iter=1000, random_state=7)
```

```python
rg.fit(X_train,Y_train)
```

```python
for col, coef in zip(X_train.columns,rg.coef_):
    print(col,coef)
```

```python
rg = Ridge(alpha=5000,random_state=7,max_iter=5000 )
rg.fit(X_train,Y_train)
for col, coef in zip(X_train.columns,rg.coef_):
    print(col,coef)
```

```python
alpha_list = [1,1000,5000]
r2_score_list=[]
for alpha_val in alpha_list:
    rg = Ridge(alpha=alpha_val,random_state=7,max_iter=5000)
    rg.fit(X_train,Y_train)
    Y_pred = rg.predict(X_test)
    r2_score_list.append(r2_score(Y_test,Y_pred))
```

```python
r2_score_list
```

# On this dataset Ridge models are performing better than Lasso model

# Conclusions from this study ::

- We need to apply power transform to convert skewed data to normal distribution

- In lasso, when we increase value of alpha, more number of features are dropped


- In ridge when we increase value of alpha then coefficient values are made smaller


# Hyper Parameter Tuning


We will be using Grid Search CV

It does grid search on given parameters and also performs cross validation

Then returns best parameter combination

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'alpha': [0.01,0.5,1,5],
    'max_iter':[100,500,1000,5000]
}
```

```python
gscv = GridSearchCV(Ridge(random_state=7),param_grid,
                scoring='r2',cv=2,verbose=2)
```

```python
gscv.fit(X_train,Y_train)
```

```python
gscv.best_score_
```

```python
gscv.best_params_
```
